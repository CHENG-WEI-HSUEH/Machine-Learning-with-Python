# Leaky ReLU Activation Function

## Applied Field

Leaky ReLU (Leaky Rectified Linear Unit) is commonly used in:

* Deep Learning
* Convolutional Neural Networks (CNNs)
* Computer Vision
* Natural Language Processing (NLP)

## Purpose

Leaky ReLU addresses the "dying ReLU" problem by allowing a small, non-zero gradient when the unit is not active.

* It prevents neurons from becoming inactive and helps maintain learning throughout the model.

## Mathematical Formula

$$
\text{LeakyReLU}(x) = 
\begin{cases}
    x, & \text{if } x \geq 0 \\
    \alpha \cdot x, & \text{if } x < 0
\end{cases}
$$

Where \$\alpha\$ is a small constant (e.g., 0.01).

### Derivative

$$
\frac{d}{dx}\text{LeakyReLU}(x) = 
\begin{cases}
    1, & \text{if } x \geq 0 \\
    \alpha, & \text{if } x < 0
\end{cases}
$$

## Python Implementation

````python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x >= 0, 1, alpha)

# Plotting Leaky ReLU and its derivative
x = np.linspace(-10, 10, 400)
y = leaky_relu(x)
dy = leaky_relu_derivative(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Leaky ReLU', color='blue')
plt.plot(x, dy, label='Derivative', linestyle='--', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title('Leaky ReLU and Its Derivative')
plt.xlabel('Input')
plt.ylabel('Output / Gradient')
plt.savefig('leaky_relu_plot.png')
plt.show()
````

````python


## C++ Implementation

```cpp
#include <vector>
#include <algorithm>

std::vector<double> leaky_relu(const std::vector<double>& x, double alpha = 0.01) {
    std::vector<double> result;
    result.reserve(x.size());
    for (double val : x) {
        result.push_back(val >= 0 ? val : alpha * val);
    }
    return result;
}

std::vector<double> leaky_relu_derivative(const std::vector<double>& x, double alpha = 0.01) {
    std::vector<double> result;
    result.reserve(x.size());
    for (double val : x) {
        result.push_back(val >= 0 ? 1.0 : alpha);
    }
    return result;
}
```

## Summary

* **Advantage**: Avoids zero gradient in negative input range.
* **Disadvantage**: The slope \$\alpha\$ is a fixed hyperparameter.
* **Common Alternatives**: Parametric ReLU, ELU, Swish.

---
````
