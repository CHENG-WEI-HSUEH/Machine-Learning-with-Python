# Tanh Activation Function: Concept, Application, and Implementation

## 1. Applied Field and Purpose

**Applied Field:**

* Machine Learning
* Neural Networks
* Deep Learning

**Purpose:**
The Tanh activation function transforms real-valued numbers into the range (-1, 1). It introduces non-linearity into neural networks and helps in centering the data, which often leads to faster convergence during training.


## 2. Mathematical Formula

* **Tanh Activation Function:**

$$
    \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

* **Tanh Derivative:**

$$
    \frac{d}{dx} \tanh(x) = 1 - \tanh^2(x)
$$

## 3. Graph Example (Python)

````python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 200)
y = tanh(x)

plt.plot(x, y)
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 200)
y = 1 - np.tanh(x) ** 2

plt.plot(x, y)
plt.title("Tanh Derivative Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
plt.show()
````

## 4. Python Implementation Example

````python
import numpy as np

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2

# Example usage
x = np.array([0.5, 1.0, -1.0])
print("Tanh:", tanh(x))
print("Tanh Derivative:", tanh_derivative(x))
````

## 5. C++ Implementation Example
```cpp
#include <iostream>
#include <cmath>

double tanh_function(double x) 
{
    return std::tanh(x);
}

double tanh_derivative(double x) 
{
    double t = std::tanh(x);
    return 1.0 - t * t;
}

int main() {
    double x = 0.5;
    std::cout << "Tanh: " << tanh_function(x) << std::endl;
    std::cout << "Tanh Derivative: " << tanh_derivative(x) << std::endl;
    return 0;
}
```

## 6. Summary

* **Tanh Function** maps input values into the (-1, 1) range.
* It is often preferred over Sigmoid because it centers the output at zero.
* The derivative formula is critical for gradient calculations in neural network training.
* Python and C++ implementations provide examples of both the function and its derivative.
* Visualizing the function helps understand its saturation behavior for large input values.
