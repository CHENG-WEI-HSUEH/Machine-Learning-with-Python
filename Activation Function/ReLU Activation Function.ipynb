## 1. Applied Field

**Purpose:**

ReLU (Rectified Linear Unit) introduces non-linearity in neural networks while maintaining computational efficiency. It solves vanishing gradient problems common in Sigmoid or Tanh.

## 2. Mathematical Formula

- **ReLU Function:**

$$
    \text{ReLU}(x) = \max(0, x)
$$

- **ReLU Derivative:**

$$
    \frac{d}{dx} \text{ReLU}(x) =
    \begin{cases}
      1, & \text{if } x > 0 \\
      0, & \text{if } x \leq 0
    \end{cases}
$$


## 3. Graph Example (Python Code)

````python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 200)
y_relu = relu(x)

plt.plot(x, y_relu)
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)
plt.show()
````

## 5. C++ Implementation Example
```c++
#include <iostream>
#include <algorithm>

double relu(double x) {
    return std::max(0.0, x);
}

double relu_derivative(double x) {
    return (x > 0) ? 1.0 : 0.0;
}

int main() {
    double x_values[] = {-2.0, -1.0, 0.0, 1.0, 2.0};
    for (double x : x_values) {
        std::cout << "ReLU(" << x << ") = " << relu(x)
                  << ", Derivative = " << relu_derivative(x) << std::endl;
    }
    return 0;
}
```

6. Summary
ReLU outputs [0,∞) range, zeroing out negative values.

- Derivative is simple: 1 for positive inputs, 0 for non-positive inputs.
- Compared to Sigmoid/Tanh, ReLU is computationally cheaper and helps deeper networks learn faster.
- Widely adopted as the default activation function in modern deep learning architectures.
