# Softmax Function

## Applied Field

Softmax is widely used in:

* Multi-Class Classification
* Deep Learning
* Natural Language Processing (NLP)
* Reinforcement Learning

## Purpose

Softmax converts a vector of values into probabilities, where the probabilities sum to 1.

* It is commonly used in the output layer of classification models.

## Mathematical Formula

Given a vector \$\mathbf{z} = \[z\_1, z\_2, \dots, z\_n]\$, the Softmax function is:

$$
\text{Softmax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{n} e^{z_j}}
$$

### Derivative

The derivative (Jacobian matrix) for the Softmax function \$S\$ is:

$$
\frac{\partial S_i}{\partial z_j} = S_i (\delta_{ij} - S_j)
$$

Where \$\delta\_{ij}\$ is the Kronecker delta.

## Python Implementation

````python
import numpy as np
import matplotlib.pyplot as plt

def softmax(z):
    exp_z = np.exp(z - np.max(z))  # for numerical stability
    return exp_z / exp_z.sum(axis=0)

# Plotting Softmax on a sample input
x = np.linspace(-2, 2, 100)
z = np.vstack([x, x + 0.5, x - 0.5])
y = np.apply_along_axis(softmax, 0, z)

plt.figure(figsize=(8, 6))
plt.plot(x, y.T)
plt.grid(True)
plt.title('Softmax Function Output Example')
plt.xlabel('Input Value')
plt.ylabel('Probability')
plt.legend(['Class 1', 'Class 2', 'Class 3'])
plt.show()
````

## C++ Implementation

```cpp
#include <vector>
#include <cmath>
#include <numeric>
using namespace std;
vector<double> softmax(const vector<double>& z) 
{
    vector<double> exp_z(z.size());
    double max_z = *max_element(z.begin(), z.end());
    double sum_exp = 0.0;

    for (size_t i = 0; i < z.size(); ++i) 
    {
        exp_z[i] = exp(z[i] - max_z);
        sum_exp += exp_z[i];
    }

    for (double &val : exp_z) 
        val /= sum_exp;
    

    return exp_z;
}
```

## Summary

* **Advantage**: Converts arbitrary scores into probabilities.
* **Disadvantage**: Can cause vanishing gradient when probabilities become extremely small.
* **Common Alternatives**: Sigmoid (for binary classification), Logistic Function.
